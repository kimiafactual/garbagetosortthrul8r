```python
dupsdf= pd.merge(uuidDF, dfpayloads, on=['factual_id', 'task_id'], how='right')
unsuredf = dupsdf[dupsdf.annotation =='unsure'] 
dupdf = dupsdf[dupsdf.annotation =='duplicate'] 
nondupdf = dupsdf[dupsdf.annotation =='non-duplicate'] 


unsure = unsuredf.groupby('payload.canonical_uuid')['factual_id'].apply(list).to_frame().reset_index()
unsure['factual_id'] = unsure['factual_id'].apply(lambda x: list(set(x)))
unsure['unsure'] = unsure['factual_id']
del unsure['factual_id']
print(len(unsure))



dupdf = dupdf.groupby('payload.canonical_uuid')['factual_id'].apply(list).to_frame().reset_index()
dupdf['factual_id'] = dupdf['factual_id'].apply(lambda x: list(set(x)))
dupdf['dupes'] = dupdf['factual_id']
del dupdf['factual_id']
print(len(dupdf))

total = pd.merge(unsure, dupdf, on ='payload.canonical_uuid',how = 'right')

nondupdf = nondupdf.groupby('payload.canonical_uuid')['factual_id'].apply(list).to_frame().reset_index()
nondupdf['factual_id'] = nondupdf['factual_id'].apply(lambda x: list(set(x)))
nondupdf['nonDupes'] = nondupdf['factual_id']
del nondupdf['factual_id']
print(len(nondupdf))

total = pd.merge(total, nondupdf, on ='payload.canonical_uuid', how = 'right')
total = total.rename(index=str, columns={'payload.canonical_uuid': 'canonical_uuid'})



total['dupes'] = total['dupes'].apply(lambda d: d if isinstance(d, list) else [])
total['unsure'] = total['unsure'].apply(lambda d: d if isinstance(d, list) else [])
total['nonDupes'] = total['nonDupes'].apply(lambda d: d if isinstance(d, list) else [])

len(total['canonical_uuid'].unique())

total2 = total.to_dict('records')


import json
with open('dups_for_ingestion.json', 'w') as outfile:
    json.dump(total2, outfile)
    
```
