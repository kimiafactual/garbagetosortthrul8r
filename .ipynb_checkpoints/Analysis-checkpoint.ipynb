```python
## yenny's file 
json_file = "duplicates_batch_2018-05-18.json"


df = pd.read_json(json_file, lines=True)
dfsub = pd.DataFrame(dict(df.submission[0:])).T
dftask = pd.DataFrame(dict(df.task[0:])).T


dfpayloads = pd.DataFrame()

# I know this is poor quality, I tried using a function (in notes)
for i in range(dfsub.shape[0]):
    f = pd.DataFrame(dfsub.payload[i]).T
    f['id'] = dfsub.id[i]
    f['task_id'] = dfsub.task_id[i]
    f['user_id'] = dfsub.user_id[i]
    f['created_at'] = dfsub.created_at[i]
    f['updated_at'] = dfsub.updated_at[i]
    dfpayloads = pd.concat([dfpayloads, f])
    
dfpayloads = dfpayloads.reset_index()
dfpayloads = dfpayloads.rename(index=str, columns={"index": "payload"})
dfpayloads['count'] = 1 # this is to signify one input per user
dfpayloads.to_csv('dfpayloads.csv')
dfpayloads['count'] = 1
```

```python
## How often is this person the only person who disagrees?
lowagreement = dfpayloads.groupby(['payload', 'annotation']).size().to_frame(name = 'count').reset_index()
lowagreement = lowagreement[lowagreement['count'] == 1]
result = pd.merge(lowagreement,dfpayloads,  how = 'right', on=['payload', 'annotation'])
result['count'] = result['count_x']

differ_from_others = (result.groupby('user_id').count()['count'].to_frame()).reset_index()
differ_from_others.sort_values('count', ascending=False)

## total 
dfpayloads['count'] = 1
totalnumber = (dfpayloads.groupby('user_id').count()['count'].to_frame()).reset_index()

final_perc = pd.merge(totalnumber,differ_from_others, how = 'left', on='user_id' )
final_perc['%'] = final_perc.count_y/final_perc.count_x
final_perc = final_perc.rename(columns={'count_y': 'Num.Divergence', 'count_x': 'TotalQs'})
print("Percent of times the user is the *only* one who disagrees")
final_perc.sort_values('%', ascending=False)

# Check to ensure we arent missing user_ids
#print(len(final_perc.user_id.unique()), len(dfpayloads.user_id.unique()))
```

```python
(final_perc.groupby('user_id').sum()['TotalQs']).plot(kind='bar')
plt.title('# of Tasks done by each Taskus')
plt.show()
```
