```python
import numpy as np
from IPython.core.display import HTML
import json
import pandas as pd
import warnings
import pandas as pd
import matplotlib as m
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize
import pickle


# # # # # # # # # # # # # # # # # # 
# This script prints out the UUIDs 
# that *were* resolved, so that they 
# can be fed into pyspark to ID whether 
# they were in edge or stable
# --> this was done for two files as an example
# --> provides a csv and a list of UUIDs
# # # # # # # # # # # # # # # # # # 

json_file = "/Users/kimia/Downloads/CA_output_1/part-00001-965c50f8-4eb3-46f1-be56-249d4dc92833-c000.json"
df = pd.read_json(json_file, lines=True)

dic = {}
for i, v in enumerate(df['_2']):
    
    js = (json.loads(df['_2'].values[i]))#['extraction']
    dic.update(js)
s = pd.Series(dic,  name='uuid')
s.index.name = 'Date'
uuids1 =((((pd.DataFrame(s)).reset_index()))['Date'])


### UUIDs that need to go into the spark job to 
### ID whether they are in edge or stable
pd.DataFrame(uuids1).to_csv('uuids1.csv')
with open("uuids1.txt", "wb") as fp:   #Pickling
    pickle.dump(uuids1, fp)

    
###############################################################
#Second file of UUIDs
###############################################################
json_file = "/Users/kimia/Downloads/CA_output_1/part-00000-965c50f8-4eb3-46f1-be56-249d4dc92833-c000.json"
df = pd.read_json(json_file, lines=True)

dic = {}
for i, v in enumerate(df['_2']):
    
    js = (json.loads(df['_2'].values[i]))#['extraction']
    dic.update(js)
s = pd.Series(dic, name='uuid')
s.index.name = 'Date'
uuids0 = ((((pd.DataFrame(s)).reset_index()))['Date'])


### UUIDs that need to go into the spark job to 
### ID whether they are in edge or stable
pd.DataFrame(uuids0).to_csv('uuids0.csv')

with open("uuids0.txt", "wb") as fp:   #Pickling
    pickle.dump(uuids1, fp)
```
