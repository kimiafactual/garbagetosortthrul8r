```python
## rank by topic area
df = result 
df = pd.concat([df, (pd.get_dummies(df['file_x']))], axis =1)
total = pd.DataFrame()
countries = ['ca-seed-inputs.json', 'it-seed-inputs.json',
       'us-seed-inputs.json', 'au-seed-inputs.json',
       'gb-seed-inputs.json', 'fr-seed-inputs.json']

def singleouttaskus(df, column_name):
    lowagreement = df.groupby([ 'task_id', str(column_name)]).size().to_frame(name = 'count').reset_index()
    lowagreement = lowagreement[lowagreement['count'] == 1]
    lowagreement =  pd.merge(lowagreement,df,  how = 'right', on=[ 'task_id',str(column_name)])
    differ_from_others = ((lowagreement.groupby('user_id').count()['count'].to_frame())).sort_values('count', ascending=False).reset_index()
    total = df.groupby(['user_id']).count()['task_id'].to_frame(name = 'count').reset_index()
    perfmstat = pd.merge(differ_from_others,total,  on=['user_id'])
    perfmstat[(str(column_name))] = (perfmstat['count_x']/perfmstat['count_y'])*100
    perfmstat = perfmstat[['user_id', str(column_name)]]
    return perfmstat





df = result 
total = pd.DataFrame()
for i,country in enumerate(countries):
    df = result[result.file_x == str(country)]
    for num,column_name in enumerate(['address', 'name', 'tel', 'website']):

        perfmstat1 = singleouttaskus(df, column_name)
        total = pd.concat([perfmstat1, total], axis = 0)
        total=total.fillna(0)
    total = (total.groupby('user_id').sum()).reset_index()

    totalstr = str(country[0:2]).upper()
    total[str(totalstr)] = (total['address'] + total['tel'] + total['website'] + total['tel'])/100
```
